```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("amazon.csv")
data.head(3)
```

```python
data.isnull().sum()
```

```python
data['rating_count'] = data['rating_count'].str.replace(',', '')
```

```python
data['rating_count'] = data['rating_count'].astype(float)
```

```python
data["rating_count"]
```

```python
data["rating_count"] = data["rating_count"].fillna(data["rating_count"].median())
```

```python
data["rating_count"].isnull().sum()
```

```python
data.info()
```

```python
data.describe()
```

```python
top_rated = data.sort_values(by = 'rating_count', ascending=False).head(10)
```

```python

plt.figure(figsize=(12, 6))
sns.set_style('whitegrid')

sns.barplot(data=top_rated, x='product_name', y='rating_count', hue='product_name', palette='magma', legend=False)

plt.xticks(rotation=70)
plt.title('Top 10 Most Rated Amazon Products', fontsize=16)
plt.xlabel('Product Name')
plt.ylabel('Number of Ratings')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data['rating'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Product Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
data['discount_percentage'] = pd.to_numeric(data['discount_percentage'].str.replace('%', ''), errors='coerce')
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='discount_percentage', y='rating', alpha=0.6, color='purple')
plt.title('Discount Percentage vs Product Rating')
plt.xlabel('Discount %')
plt.ylabel('Rating')
plt.tight_layout()
plt.show()
```
